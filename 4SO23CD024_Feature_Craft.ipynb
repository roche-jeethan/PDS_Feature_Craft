### Task 2:

 What  is  the  seasonal  performance  of  each  crop,  in  terms  of  both  total  Production  and
 average Yield

 Hints  :

 ●  df.groupby a list of column names
  
 ●  agg():   apply one or more aggregation functions to one or more columns,

### Task 3:

 What is the year-wise total production for each crop

 Hints  :

 ●  df.pivot(): transform a DataFrame from a "long" format to a "wide" format

 Arguments: index, columns, and values.

```python
"""
We have to understand what are we even doing. Thing is there are generally 2 ways we categorize the data

One is called Long format... where all the possibilities are given one by one in a more row fashion

Other one is called Wide format ... where the values are transformed more on the column fashion

So the process of
Long -> wide: is called Pivot
Wide -> long: is called Melting

"""

wide_format = df.pivot_table(
    index='Crop',
    columns = 'Crop_Year',
    values = 'Production',
    aggfunc = 'sum'
)
wide_format
```

### Task 4:

Which crop, in which year, produced the highest average yield

 Hints  :

 ●  df.melt(): transform or reshape DataFrames from a "wide" format to a "long"  format.

Arguments: id_vars, value_vars, var_name, value_name
 [use transformed data from pivot)

```python
year_list = df['Crop_Year'].sort_values().unique().tolist()
print(year_list)
```

```python
# But now we are doing the other way around...
long_format = pd.melt(wide_format.reset_index(), id_vars=['Crop'], value_vars=year_list, value_name='Production')
long_format
```

## 3. Cleaning

 ### Task 1:
 Change  inconsistent  columns  values  like  'State'  and  'District'  from  uppercase  to
 lowercase and remove trailing or leading spaces

 Hints  :

 ●  df[‘col’].str.lower(), df[‘col’].str.strip()

```python
df['State'] = df['State'].str.replace(' ', '_') # this one replace space with underscore
df['District'] = df['District'].str.capitalize() # this one capitalize the first letter of the word
df['District'] = df['District'].str.replace(' ', '_')
df['Crop'] = df['Crop'].str.replace(' ', '_')
```

### Task 2:
 Instead  of  filling  missing  Production  with  0  or  a  global  mean,  impute  the  NaN  values
 by  calculating  Area  *  (Average_Yield).  The  Average_Yield  should  be  the  average  for
 that specific District and Crop combination.

 Hints  :

 ●  groupby(['District', 'Crop']), transform('mean'), and fillna()).

```python
df.head()
```

```python
# df['Production'].isna().sum()
```

```python
# i have found around 4935 null values in production column, lets fill with something else other than mean
average_yield = df.groupby(['District','Crop'])['Yield'].transform('mean')

# Fill NaN values in 'Production' by multiplying 'Area' with the calculated average yield
df['Production'] = df['Production'].fillna(df['Area'] * average_yield)
```
